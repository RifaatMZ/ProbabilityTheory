#### 1- Провести дисперсионный анализ для определения того, есть ли различия среднего роста среди взрослых футболистов, хоккеистов и штангистов. Даны значения роста в трех группах случайно выбранных спортсменов: Футболисты: 173, 175, 180, 178, 177, 185, 183, 182. Хоккеисты: 177, 179, 180, 188, 177, 172, 171, 184, 180. Штангисты: 172, 173, 169, 177, 166, 180, 178, 177, 172, 166, 170.

```python
ft = np.array([173, 175, 180, 178, 177, 185, 183, 182])
hk = np.array([177, 179, 180, 188, 177, 172, 171, 184, 180])
pl = np.array([172, 173, 169, 177, 166, 180, 178, 177, 172, 166, 170])
```

```python
# метод ANOVA 
from scipy import stats
# stats.f_oneway?
```

```python
y1 = ft
y2 = hk
y3 = pl
```

```python
stats.f_oneway(y1, y2, y3)
```

```python
# уровнь значимость мы выбрали 0.05
# pvalue < 0.05
# Следовательно, различие между группами статистически значимое
```

```python
# второй способ
k = 3
n1 = ft.size
n2 = hk.size
n3 = pl.size
n = n1 + n2 + n3
n1, n2, n3, n
```

```python
y1_mean = y1.mean()
y2_mean = y2.mean()
y3_mean = y3.mean()
y1_mean, y2_mean, y3_mean
```

```python
y_mean = (np.sum(y1) + np.sum(y2) + np.sum(y3)) / n
y_mean
```

```python
S = np.sum((y1 - y_mean) ** 2 ) + np.sum((y2 - y_mean) ** 2 ) + np.sum((y3 - y_mean) ** 2 )
S
```

```python
Sf = ((np.sum((y1_mean - y_mean) ** 2)) * n1) + ((np.sum((y2_mean - y_mean) ** 2)) * n2) + \
    ((np.sum((y3_mean - y_mean) ** 2)) * n3)
Sf
```

```python
Soct = np.sum((y1 - y1_mean) ** 2 ) + np.sum((y2 - y2_mean) ** 2 ) + np.sum((y3 - y3_mean) ** 2 )
Soct
```

```python
S_ = Sf + Soct
S_
```

```python
# фактирная дисперсия
af = Sf / (k - 1)
af
```

```python
# остаточняя дисперсия
aoct = Soct / (n - k)
aoct
```

```python
Fh = af / aoct
Fh
```

```python
# найдем Fкрит из таблицы критических точек распределения фишера-снедекора 
# для уровня значимости 0.05 (k -1 = 2, n - k = 25)
```

```python
Fkpt = 2.39
```

```python
Fh > Fkpt
```

```python
# Следовательно, различие между группами статистически значимое
```
